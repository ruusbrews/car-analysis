### Find the car (or cars) with the best ratio of horsepower to weight among all cars that also have above-median mpg.

```python
# filter out above-median mpg
median_mpg = data['mpg'].median()
above_median = data[data['mpg'] > median_mpg].copy()

# calc horsepower to weight ratio
above_median['hp_to_w'] = above_median['horsepower'] / above_median['weight']

# find cars with highest ratio
highest_ratio = above_median['hp_to_w'].max()
best_cars = above_median[above_median['hp_to_w'] == highest_ratio]

best_cars['name']
```

### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

```python
avg_mpg = data.groupby(['model_year', 'origin'])['mpg'].mean().reset_index()

plt.figure(figsize=(10,6))
sns.lineplot(data=avg_mpg, x='model_year', y='mpg', hue='origin', marker='x')
plt.title('Average MPG Over the Years by Origin')
plt.ylabel('Average MPG')
plt.xlabel('Model Year')
plt.legend(title='Origin')

plt.show()
```

### We define a “consistent” car model as one that was produced over multiple years and had very low variation in mpg across those years (standard deviation < 1.0).

Tasks:

Identify car names that appear in more than one model_year.

For each such name, compute the standard deviation of mpg across years.

Return the car(s) with the lowest variation in mpg, among those with at least 2 appearances and std(mpg) < 1.0.

Report the model name(s), number of appearances, and the average mpg.

Bonus:
Sort the result by number of appearances (descending), then mpg (descending).

```python
# identifying cars in more than one year + getting this data as separate dataset
car_count = data.groupby('name')['model_year'].nunique()
multiple_years = car_count[car_count > 1].index
multiple_years_data = data[data['name'].isin(multiple_years)]

# compute standard dev of mpg across years for those cars
sd_calculate = multiple_years_data.groupby('name')['mpg'].agg(['std','count','mean'])

# filter out copy of dataset with only consistent cars w lowest variation in mpg AND std(mpg)<1
consistent_cars = sd_calculate[sd_calculate['std'] < 1.0].copy()

# sort by appearance descending then mpg descending
consistent_cars = consistent_cars.sort_values(by=['count','mean'], ascending=[False,False])
consistent_cars
```
